# storing data in chromadb

```python
from langchain.vectorstores import Chroma
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain.docstore.document import Document
import os

# Set up environment
os.environ["GOOGLE_API_KEY"] = "AIzaSyDK6TIx35vr_UylDTP9V65CIfth0pqFozg"  # or use dotenv

# Initialize embeddings
embedding_function = GoogleGenerativeAIEmbeddings(model="models/embedding-001")
```

```python
import uuid
```

```python
document_id = str(uuid.uuid4())
filename = "test"
```

```python

# Example text to store
text = """
Artificial Intelligence (AI) is revolutionizing industries by automating tasks, enabling data-driven decisions, 
and enhancing human capabilities. Natural Language Processing (NLP) is a key area of AI that deals with understanding 
and generating human language.
"""

# Convert to LangChain Document
documents = [Document(page_content=text, metadata={"source": "example_doc"})]

# Split the document into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=10)
docs = text_splitter.split_documents(documents)


# âœ… Add metadata for filtering later
for doc in docs:
    doc.metadata["document_id"] = document_id
    doc.metadata["filename"] = filename


# Create ChromaDB and persist
persist_dir = "./chroma_db_store"
vectordb = Chroma.from_documents(documents=docs, embedding=embedding_function, persist_directory=persist_dir)
vectordb.persist()
```
